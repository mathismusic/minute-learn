```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_linearly_separable(n=100, dim=2, seed=None):
    """
    Generate a linearly separable dataset
    """
    if seed is not None:
        gen = np.random.default_rng(seed)
    weights = 2 * gen.random(dim) - 1
    bias = 2 * gen.random() - 1
    X = 10 * gen.random((n, dim)) - 5
    y = (X.dot(weights) + bias > 0).astype(int)
    return X, y
```

```python
X, y = generate_linearly_separable(n=100, dim=2, seed=42)
plt.scatter(X[:, 0], X[:, 1], c=y, marker='x', cmap='bwr')
```

```python
def generate_regression(n=100, dim=2, sigma=1, seed=None):
    """
    Generate a regression dataset
    """
    if seed is not None:
        gen = np.random.default_rng(seed)
    weights = 2 * gen.random(dim) - 1
    bias = 2 * gen.random() - 1
    X = 10 * gen.random((n, dim)) - 5
    y = X.dot(weights) + bias + sigma * gen.standard_normal(n)
    return X, y
```

```python
X, y = generate_regression(n=1000, dim=1, sigma=0.1, seed=42)
plt.plot(X, y, 'x')
```
